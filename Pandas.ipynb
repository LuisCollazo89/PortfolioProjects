# Group By

```python
df3 = pd.read_csv(r"C:\Users\Luis\Desktop\Python\DataAnalysis\TheAnalyst\Data\Flavors.csv")
```

```python
df3
```

```python
group_by_frame = df3.groupby("Base Flavor")
```

```python
group_by_frame.mean(numeric_only = True)
```

```python
df3.groupby("Base Flavor").count()
```

```python
df3.groupby("Base Flavor").min()
```

```python
df3.groupby("Base Flavor").max()
```

```python
df3.groupby("Base Flavor").sum(numeric_only = True)
```

```python
df3.groupby("Base Flavor").agg({"Flavor Rating":["mean", "max","count","sum"]})
```

```python
df3.groupby("Base Flavor").agg({"Flavor Rating":["mean", "max","count","sum"],"Texture Rating":["mean", "max","count","sum"]})
```

```python
df3.groupby(["Base Flavor", "Liked"]).mean(numeric_only = True)
```

# Merging, Joining

```python
df4 = pd.read_csv(r"C:\Users\Luis\Desktop\Python\DataAnalysis\TheAnalyst\Data\LOTR.csv")
```

```python
df5 = pd.read_csv(r"C:\Users\Luis\Desktop\Python\DataAnalysis\TheAnalyst\Data\LOTR 2.csv")
```

```python
df4.merge(df5)
```

```python
df4.merge(df5, "inner", on ="FellowshipID")
```

# Outer

```python
df4.merge(df5, "outer")
```

```python
df4.merge(df5, "outer", on ="FellowshipID")
```

# Left Join


```python
df4.merge(df5, "left")
```

# Concat

```python
pd.concat([df4,df5])
```

```python
pd.concat([df4,df5],join ="outer", axis = 1)
```
